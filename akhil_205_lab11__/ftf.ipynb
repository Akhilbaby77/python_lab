```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python



data = pd.read_csv("./train.csv")


df = pd.DataFrame(data)
df = df.head(50)
print(df)

plt.plot(df['Hour'], df['Temperature'])
plt.xlabel('Hour')
plt.ylabel('Temperature')
plt.title('Line Graph of Hour vs. Temperature')
plt.show()


plt.bar(df['Hour'], df['Temperature'])
plt.xlabel('Hour')
plt.ylabel('Temperature')
plt.title('Bar Graph of Hour vs. Temperature')
plt.show()


plt.scatter(df['Hour'], df['Temperature'])
plt.xlabel('Hour')
plt.ylabel('Temperature')
plt.title('Scatter Plot of Hour vs. Temperature')
plt.show()


```
